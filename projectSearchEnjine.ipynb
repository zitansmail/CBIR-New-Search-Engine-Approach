```python
def show_values(axs, orient="v", space=.05):
    def _single(ax):
        if orient == "v":
            for p in ax.patches:
                _x = p.get_x() + p.get_width() / 2
                _y = p.get_y() + p.get_height() + (p.get_height()*0.01)
                value = '{:.1f}'.format(p.get_height())
                ax.text(_x, _y, value, ha="center") 
        elif orient == "h":
            for p in ax.patches:
                _x = p.get_x() + p.get_width() + float(space)
                _y = p.get_y() + p.get_height() - (p.get_height()*0.5)
                value = '{:.1f}'.format(p.get_width())
                ax.text(_x, _y, value, ha="left")

    if isinstance(axs, np.ndarray):
        for idx, ax in np.ndenumerate(axs):
            _single(ax)
    else:
        _single(axs)
```

```python
def get_similarity_between_two_senrences(X,Y):
    X = X.lower()
    Y = Y.lower()
    # tokenization
    X_list = word_tokenize(X) 
    Y_list = word_tokenize(Y)
  
    #print(X_list,Y_list)

    # sw contains the list of stopwords
    sw = stopwords.words('english') 
    l1 =[];l2 =[]
  
    # remove stop words from the string
    X_set = {w for w in X_list if not w in sw} 
    Y_set = {w for w in Y_list if not w in sw}

    #print(X_set,Y_set)
  
    # form a set containing keywords of both strings 
    rvector = X_set.union(Y_set) 

    #print(rvector)


    for w in rvector:
        if w in X_set: l1.append(1) # create a vector
        else: l1.append(0)
        if w in Y_set: l2.append(1)
        else: l2.append(0)
    #print(X_set,Y_set)
    #print(l1,l2)
    c = 0
    #print(sum(l1),sum(l2))  
    # cosine formula 
    for i in range(len(rvector)):
        c+= l1[i]*l2[i]
    #print(c)
    cosine = c / float((sum(l1)*sum(l2))**0.5)
    print("similarity: ", cosine)
    return cosine
```

```python
def unique_values(array):
    pre_ids = []
    ids = []
    for i in array:
        if int(i) not in pre_ids:
            pre_ids.append(int(i))
            ids.append(int(i))
    return ids
def find_similar_offers(ids_images,id_query):
    similar_ids = []
    for id_offer in ids_images:
        if id_offer is not id_query:
            if get_similarity_between_two_senrences(get_all_infos_from_db(id_query,'agency_name'),get_all_infos_from_db(id_offer,'agency_name')) > 0.5:
                if get_all_infos_from_db(id_query,'city') == get_all_infos_from_db(id_offer,'city') and get_all_infos_from_db(id_query,'price') == get_all_infos_from_db(id_offer,'price'):
                    similar_ids.append(id_offer)
                #print(get_all_infos_from_db(id_query,'price'))
                #print(get_all_infos_from_db(id_query,'id'))
                #print(get_all_infos_from_db(id_offer,'price'))
                #print(get_all_infos_from_db(id_offer,'id'))
    similar_ids.append(id_query)
    return unique_values(similar_ids)
        
```

```python
def str_to_vec(str1,str2):
    edits = [[x for x in range(len(str1) + 1)] for y in range(len(str2)+ 1)]
    print(edits)
    for i in range(len(str2) + 1):
        edits[i][0] = i
    for i in range(1, len(str2) + 1):
        for j in range(1,  len(str1) + 1):
            if str2[i-1] == str1[j-1]:
                edits[i][j] = edits[i-1][j-1]
            else:
                edits[i][j] = 1 + min(edits[i-1][j-1], edits[i-1][j],
                                     edits[i][j-1])
    return edits[-1][-1]
```
